```python
# read tesla.csv and print the first 5 rows
import pandas as pd
df = pd.read_csv('tesla.csv')
print(df.head(5))
```

```python
# plot the price of tesla with dates on x-axis

import matplotlib.pyplot as plt 
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['Price'].plot()
plt.show()
```

```python
# show prices for each year. x-axis should be months and y-axis should be price
df['Year'] = df.index.year
df['Month'] = df.index.month
df.groupby('Year')['Price'].plot()
plt.show()
```

```python
# whats is average rate increase for each month
df['Price_diff'] = df['Price'].diff()
df.groupby('Month')['Price_diff'].mean().plot(kind='bar')
plt.show()
```

```python
# Show all prices for each year for each month color coded
df.groupby(['Year','Month'])['Price'].mean().unstack().plot()
plt.show()
```
